## Description of the dataset

In this part of the assignment we will explore the MNIST digits dataset. We expect the digits to lie in a lower-dimensional manifold and want to examine the representation we get by applying Principal Components Analysis (PCA). PCA maps the data into a new space by effectively rotating the base vectors of the input space to the directions with the highest variance.

### Description of the Data

This MNIST Dataset is a collection of handwritten digits. The samples are partitioned (nearly) evenly across the 10 different digit classes {0, 1, . . . , 9}. We use a preprocessed version for which the data are $8 \times 8$ pixel images containing one digit each. For further details on how the digits are preprocessed, see the sklearn documentation. The images are grayscale, with each pixel taking values in {0, 1, . . . , 16}, where 0 corresponds to black (weakest intensity) and 16 corresponds to white (strongest intensity). Therefore, the dataset is a N × 64
dimensional matrix where each dimension corresponds to a pixel from the image and N is the number of
images.

Let us visualise the explained variance by the PCA components. Familiarise yourself with the [PCA](http://scikit-learn.org/0.19/modules/generated/sklearn.decomposition.PCA.html) class in SKLearn.

Set `svd_solver='full'` and keep all other parameters at default values. Indicate the point at which 80% of the variance is explained (e.g. using matplotlib's [`vlines()`](https://matplotlib.org/2.2.3/api/_as_gen/matplotlib.pyplot.vlines.html)): label the number of components at this point (e.g. in the Legend). Also, for consistency, number components starting from 0.* <br>
&nbsp;&nbsp;**(b)** [Text] By commenting on the shape of the plot, why do you think 80% of the variance seems like a reasonable threshold? *Hint: Think about why we do PCA*. (<3 sentences)<br>
&nbsp;&nbsp;**(c)** [Text] What can you say about the number of components required to explain 80% of the variance? (<4 sentences)

````python
#(a) # Your Code goes here:
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(data)
# Apply normalization on the set
data_scale = scaler.transform(data)

pca = PCA(svd_solver='full')
pca.fit(data_scale)
data_new = pca.transform(data_scale) #transform features into new feature space
var_ratios = pca.explained_variance_ratio_
var_ratios = sorted(var_ratios, reverse=True)
var_cumulative = np.cumsum(var_ratios)
# print(var_ratios)
print(var_cumulative)


fig = plt.figure()
ax1 = fig.add_subplot(111)

pylab.rcParams['figure.figsize'] = (12, 8)

x = np.arange(0,64)

ax1.bar(x, var_ratios,width=0.2,align='center', color='tab:blue')
ax1.set_xlabel("Dimension")
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.set_ylabel("Explained Variance Ratio", color='tab:blue')

ax2 = ax1.twinx()
ax2.errorbar(x, var_cumulative, color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.set_ylabel("Cumulative Variance", color='tab:red')


x_80 = next(x[0] for x in enumerate(var_cumulative) if x[1] > 0.8) #gets first index where 80% of variance is explained
plt.vlines(x_80, 0.0, 1, linestyles='dashed', label="80% Variance")
# dim_scaled = [math.log(n,10) for n in dimensionality] #log-scale the dimensionality to adjust for low values
````

(b) ***Your answer goes here:***
By commenting on the shape of the plot, why do you think 80% of the variance seems like a reasonable threshold? Hint: Think about why we do PCA. (<3 sentences)

The point of PCA is to get a stronger visualization of the data while reducing the complexity, hence, it is important for us to find the principle components which give us the greatest explanatory power(highest variance) while minimizing the number we need. The cumulative variance curve has a logarithmic shape, which means the principal component vectors on the left account for a much greater proportion of the total variance of the space than those on the right. Because of this, each principle component after the 80% cutoff contributes a very small amount of variance to our PCA model, hence it is logical to reduce our dimensionality at this point rather than to continue to add principal component vectors with little explanatory power. 

(c) ***Your answer goes here:***
What can you say about the number of components required to explain 80% of the variance?

Only 20 components are required to explain 80% of the variance of our original space, which is less than a third of the original dimensionality. This is great, since it means that the 20 principal components with the greatest variance give us a very strong visualization of the dataset while reducing the dimensionality significantly.

````python
#(a) # Your Code goes here:
digit_0 = digits.data[0,:]
fig, ax = plt.subplots(figsize=(9,7))
ax = sns.heatmap(digit_0.reshape(8,8))
plt.title("Origitnal Digit 0")

digit_0 = (digit_0.reshape(1,-1)) #reshape data
transformed_data = pca.transform(digit_0) #transformed digit, combination of pca components

new_0 = pca.inverse_transform(transformed_data) #original digit reconstructed by multiplying each dimension with projection value

fig, ax = plt.subplots(figsize=(9,7))
ax = sns.heatmap(new_0.reshape(8,8))
plt.title("Reconstructed Digit 0")

MAD = np.abs(digit_0.reshape(1,-1) - (new_0)).mean()
print("MAD between digits: ",MAD)
````
