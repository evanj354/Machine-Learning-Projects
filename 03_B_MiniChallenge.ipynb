````python
# This is where your working code should start. Fell free to add as many code-cells as necessary.
#  Make sure however that all working code cells come BEFORE the #ANSWER_TEXT# and #ANSWER_PROB#
#  cells below.

# Your Code goes here:
data_path = os.path.join(os.getcwd(), 'datasets', 'Images_C_Train.csv')
imgs_train = pd.read_csv(data_path, delimiter = ',')
data_path = os.path.join(os.getcwd(), 'datasets', 'Images_C_Validate.csv')
imgs_validate = pd.read_csv(data_path, delimiter = ',')
data_path = os.path.join(os.getcwd(), 'datasets', 'Images_C_Test.csv')
imgs_test = pd.read_csv(data_path, delimiter = ',')


imgs_train.drop(["imgId"], axis=1, inplace=True)
imgs_validate.drop(["imgId"], axis=1, inplace=True)

attributes = ["is_aeroplane","is_bicycle","is_bird","is_boat","is_bottle","is_bus","is_car","is_cat","is_chair","is_cow","is_diningtable","is_dog","is_horse","is_motorbike","is_pottedplant","is_sheep","is_sofa","is_tvmonitor"]
for attribute in attributes:
    imgs_train.drop([attribute], axis=1, inplace=True)
    imgs_validate.drop([attribute], axis=1, inplace=True)
imgs_validate.head(5)


print("Training set shape: ", imgs_train.shape)
print("Validation set shape: ", imgs_validate.shape)
print("Testing set shape: ", imgs_test.shape)


X_train = imgs_train.drop(["is_person"], axis=1)
y_train = imgs_train["is_person"]

X_validate = imgs_validate.drop(["is_person"], axis=1)
y_validate = imgs_validate["is_person"]

X_test = imgs_test.drop(["is_person"], axis=1)
y_test = imgs_test["is_person"]

outliers_train = X_train[np.sum(X_train.values, axis=1) > 1.2].index
outliers_validate = X_validate[np.sum(X_validate.values, axis=1) > 1.2].index
outliers_test = X_test[np.sum(X_test.values, axis=1) > 1.2].index

X_train.drop(outliers_train, axis=0, inplace=True)
y_train.drop(outliers_train, axis=0, inplace=True)
X_validate.drop(outliers_validate, axis=0, inplace=True)
y_validate.drop(outliers_validate, axis=0, inplace=True)
X_test.drop(outliers_test, axis=0, inplace=True)
y_test.drop(outliers_test, axis=0, inplace=True)




# for row in X_train.index.values:
#     sum = np.sum(X_train.loc[row:row].values)
#     if sum > 1.1:
#         print(sum)

# print(X_train)

scaler = StandardScaler()
# Fit on training set only.
scaler.fit(X_train)
# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_validate = scaler.transform(X_validate)
X_test = scaler.transform(X_test)
# print(X_train[0:5])


pca = PCA(.9)
pca.fit(X_train)
print(pca.n_components_)

X_train = pca.transform(X_train)
X_validate = pca.transform(X_validate)
X_test = pca.transform(X_test)

lr = LogisticRegression(solver='lbfgs', C=0.001)
lr.fit(X_train, y_train)
validate_pred_probabilities = lr.predict_proba(X_validate)
pred_probabilities = lr.predict_proba(X_test)

total_loss = log_loss(y_validate, validate_pred_probabilities)

y_pred = lr.predict(X_validate)

print("Likelihoods for each instance: \n", pred_probabilities)
print('\nLogistic Classification accuracy on validation set: {:.3f}'.format(lr.score(X_validate, y_validate)))
print("\nLog Loss: ", total_loss)
cm = confusion_matrix(y_validate, y_pred)
plot_confusion_matrix(cm=cm)


clf = svm.SVC(kernel="rbf", gamma='auto', C=1, probability=True)
clf.fit(X_train, y_train)
y_pred = clf.predict_proba(X_validate)
print('\nSVM Classification accuracy on validation set: {:.3f}'.format(clf.score(X_validate, y_validate)))
print("\nLog Loss SVM: ", log_loss(y_validate, y_pred))
````
